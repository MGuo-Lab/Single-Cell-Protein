```python
import geopandas as gpd
import pandas as pd
```

```python
#world shapefile
poly=gpd.read_file("dataset/wb_countries_admin0_10m/wb_countries_admin0_10m.shp")
```

```python
#check crs
poly.crs
```

```python
poly.columns
```

```python
from shapely import wkt
#rainfall point data 2017
rain17 = pd.read_csv('dataset/rainfall_point_17.csv')
rain17['geometry'] =rain17['geometry'].apply(wkt.loads)
g_rain17 = gpd.GeoDataFrame(rain17)
```

```python
g_rain17.crs="EPSG:3857"
```

```python
g_rain17=g_rain17.to_crs("EPSG:4326")
```

```python
from geopandas.tools import sjoin
points_polys_17 =gpd.sjoin(g_rain17, poly, how="left")
points_polys_17.head()
```

```python
points_polys_17=points_polys_17.dropna(axis=0,subset=["OBJECTID"])
```

```python
stats_pt_17 = points_polys_17.groupby('OBJECTID')['pre_anual'].agg(['mean','median','max','min']).reset_index()
```

```python
poly_name=pd.DataFrame(poly,columns=["OBJECTID","FORMAL_EN","NAME_EN"])
```

```python
rainfall_data_17=pd.merge(stats_pt_17,poly_name,on="OBJECTID")
```

```python
#export csv data 
rainfall_data_17.to_csv('dataset/rainfall_17.csv', index=False,encoding="utf-8")
```
